### 2) Resizing

To resize, we need different interpolation methods
1) cv2.INTER_AREA - Used to shrink an image<br>
2) cv2.INTER_LINEAR - Used to expand an image<br>
2) cv2.INTER_CUBIC - Can be used for both<br>    


```python
r1 = cv2.imread('car1.jpg')
r1 = cv2.cvtColor(r1,cv2.COLOR_BGR2RGB)
plt.imshow(r1)
plt.show()
```

```python
r1.shape
```

```python
d1 = cv2.resize(r1,(410,240),cv2.INTER_LINEAR)
d2 = cv2.resize(r1,(150,85),cv2.INTER_AREA)
d3 = cv2.resize(r1,(150,280),cv2.INTER_CUBIC)
d4 = cv2.resize(r1,(450,100),cv2.INTER_CUBIC)

plt.imshow(r1),plt.title('Original'),plt.show()
plt.imshow(d1),plt.title('Expanded Img'),plt.show()
plt.imshow(d2),plt.title('Shrinked img'),plt.show()
plt.imshow(d3),plt.title('d3'),plt.show()
plt.imshow(d4),plt.title('d4'),plt.show()
```

### Thresholding

Thresholidng is connverting greyscale images into its binary format

### Methods of Thresholding

Assumed threshold is t

1) <b>cv2.THRESH_BINARY</b><br>
Converts pixels below t to 0(black) and above t to 255(white).

2) <b>cv2.THRESH_BINARY_INV</b><br>
Converts pixels below t to 255(white) and above t to 0(black)

3) <b>cv2.THRESH_TRUNC</b><br>
Pixels below t remain unchanged and pixels aboe t are truncated (held at t).

4) <b>cv2.THRESH_TOZERO</b><br>  Pixels below t are converted to 0(black) and pixels above t held as it is (remain unchanged)

5) <b>cv2.THRESH_TOZERO_INV</b><br>  Pixels above t are converted to 0(black) and pixels nbelow t held as it is (remain unchanged)

```python
m = cv2.imread('kneau.jpg',0)
plt.imshow(m,'gray')
plt.show()
```

```python
print(m[30])
```

```python
t = 127
_,t1 = cv2.threshold(m,t,255,cv2.THRESH_BINARY)
plt.imshow(t1,'gray')
plt.show()
```

```python
print(t1[30])
```

```python
t = 127
_,t2 = cv2.threshold(m,t,255,cv2.THRESH_BINARY_INV)
plt.imshow(t2,'gray')
plt.show()
```

```python
print(t2[30])
```

```python
t = 127
# Pixels below t remain unchanged and pixels above t are truncated (held at t).
_,t3 = cv2.threshold(m,t,255,cv2.THRESH_TRUNC)
plt.imshow(t3,'gray')
plt.show()
```

```python
print(t3[30])
```

```python
t = 127
# Pixels below t are converted to 0(black) and pixels above t held as it is (remain unchanged)
_,t4 = cv2.threshold(m,t,255,cv2.THRESH_TOZERO)
plt.imshow(t4,'gray')
plt.show()
```

```python
print(t4[30])
```

```python
t = 127
# Pixels above t are converted to 0(black) and pixels below t held as it is (remain unchanged)
_,t5 = cv2.threshold(m,t,255,cv2.THRESH_TOZERO_INV)
plt.imshow(t5,'gray')
plt.show()
```

```python
print(t5[30])
```

#### Image Segmentation

```python
from sklearn.cluster import KMeans
```

```python
s1 = cv2.imread('s1.jpg')
s1 = cv2.cvtColor(s1,cv2.COLOR_BGR2RGB)
plt.imshow(s1)
plt.show()
```

```python
def rgb2hex(rgb):
    hex_img = "#{:02x}{:02x}{:02x}".format(int(rgb[0]),int(rgb[1]),int(rgb[2]))
    return hex_img

print(rgb2hex([255,0,0]))
print(rgb2hex([0,0,255]))
print(rgb2hex([0,255,0]))
```

```python
s1.shape
```

```python
s1_re = s1.reshape(183*275,3)
print(s1_re.shape)
```

```python
k5 = KMeans(n_clusters=5)
k5.fit(s1_re)
```

```python
ypred5 = k5.predict(s1_re)
```

```python
from collections import Counter
```

```python
c1 = Counter(ypred5)
print(c1)
```

```python
print(sum(c1.values()))
```

```python
centers = k5.cluster_centers_
print(centers)
```

```python
col_in_img = [rgb2hex(i) for i in centers]
print(col_in_img)
```

```python
print("{}-{}".format(c1.keys(),col_in_img))
```

```python
plt.pie(c1.values(),labels=col_in_img,colors=col_in_img,autopct='%.2f%%')
plt.show()
```

```python
def img_seg(img,k):
    img_re = img.reshape(img.shape[0]*img.shape[1],img.shape[2])
    m = KMeans(n_clusters=k)
    m.fit(img_re)
    ypred = m.predict(img_re)
    c1 = Counter(ypred)
    cent = m.cluster_centers_
    col_in_img = [rgb2hex(i) for i in cent]
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,10))
    ax1.imshow(img)
    ax1.set_title('Original img')
    
    ax2.pie(c1.values(),labels=col_in_img,colors=col_in_img,autopct="%.2f%%")
    ax2.set_title(f'Segmented image into {k} clusters')
    plt.show()
```

```python
img_seg(s1,7)
```

```python
img_seg(w1,3)
```
